```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,10,25)
y=x**3
print (x)
print (y)
```

```python
# plotting created x and y variables above using "plt.plot"
plt.plot(x,y,'purple',linewidth=2,linestyle='-') # providing the color, width and line style as input.
plt.xlabel("time")
plt.ylabel("number of iteration")
plt.title("number of iteration Vs time")
```

```python
#creating hte sub plot of three plots as illustrated
fig=plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(x,y,"b")
plt.title('plot 1')
plt.subplot(1,3,2)
plt.plot(y*2,x**2,'r')
plt.title('plot2')
plt.subplot(1,3,3)
plt.plot(x,y,'yellow')
plt.title('plot3')
```

```python
# plotting a plot inside another plot
fig=plt.figure()
fig=plt.figure(figsize=(10,6))# provide the dimension of the plot as input
fig1=fig.add_axes([0.1,0.1,0.8,0.8])
fig2=fig.add_axes([0.25,0.5,0.3,0.2])
fig1.plot(x,y,'black')
fig1.set_title('large plot')
fig1.set_xlabel('time')
fig1.set_ylabel('number of iteration')
fig2.plot(x*2,y)
fig2.set_title('small plot')
fig2.set_xlabel('time')
fig2.set_ylabel('number of iteration')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(5,5),dpi=100)
axes[0,0].plot(x,y,'aqua')
axes[0,0].set_title('plot title')
axes[0,0].set_xlabel('time')
axes[0,0].set_ylabel('production')
#plot 2
axes[0,1].plot(y,x,'pink')
axes[0,1].set_title('plot2')
axes[0,1].set_xlabel('time')
axes[0,1].set_ylabel('production 2')
#plot 3
axes[1,0].plot(y,x**2,'green')
axes[1,0].set_title('plot3')
axes[1,0].set_xlabel('time')
axes[1,0].set_ylabel('production 3')
# plot 4
axes[1,1].plot(x,y**2,'brown')
axes[1,1].set_title('plot 4')
axes[1,1].set_xlabel('time')
axes[1,1].set_ylabel('production 4')
plt.tight_layout()
```
